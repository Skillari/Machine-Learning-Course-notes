```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# lets create the dataset 

```python
from sklearn.datasets import make_blobs
```

```python
X ,y = make_blobs(n_samples= 600 ,centers = 6 , random_state = 101 , cluster_std=0.70)
```

```python
X
```

```python
plt.scatter(X[:,0] ,X[:,1],s =10 )
```

# fitting our model 

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
```

```python
y_kmeans
```

```python
plt.scatter(X[:,0] ,X[:,1],s =10 ,c = y_kmeans )
```

```python
centers  = kmeans.cluster_centers_
```

```python
centers
```

```python
plt.scatter(X[:,0] ,X[:,1],s =10 ,c = y_kmeans )
plt.scatter(centers[:,0] ,centers[:,1],s =200 ,c = 'black' , alpha = 0.5 )
```

# elbow method 

```python
dist = []
for k in range(1,15) :
    model = KMeans(k)
    model.fit(X)
    dist.append(model.inertia_)
```

```python
dist
```

```python
plt.plot(range(1,len(dist)+1),dist)
plt.title("Elbow method")
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
```

optimal value of k = 6

# retrain our model 

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
centers  = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0] ,X[:,1],s =10 ,c = y_kmeans )
plt.scatter(centers[:,0] ,centers[:,1],s =200 ,c = 'black' , alpha = 0.5 )
```
